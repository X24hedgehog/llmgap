Check how different tokenizers handle numbers

```python
import os
os.environ["HF_TOKEN"] = ""
```

```python
from transformers import AutoTokenizer
import matplotlib.pyplot as plt
import numpy as np

model_ids = ["meta-llama/Llama-3.1-8B-Instruct", "meta-llama/Llama-3.1-70B-Instruct", "Qwen/Qwen2.5-7B-Instruct", "meta-llama/Llama-2-7b-chat-hf"]


for model_id in model_ids:
    tokenizer = AutoTokenizer.from_pretrained(model_id)

    tkns_by_len = [[], [], [], [], [], []]
    for i in range(20000):
        tokens = tokenizer.tokenize(str(i))
        tkns_by_len[len(str(i))].append(len(tokens))
    
    xs = list(range(1,6))
    plt.plot(xs, [np.mean(tkns_by_len[i]) for i in xs], label=model_id, marker="o", alpha=0.5)
    plt.legend()
```
